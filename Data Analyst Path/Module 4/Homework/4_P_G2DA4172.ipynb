### Number 5

Let's say we have loop function to calculate sum of all numbers from 1 to a given number. What is the result of sum if we input 100? 

If we input number 5, we will sum 1+2+3+4+5=15, so the result will be 15.

```python
def sum_number (number):
  result = []
  for i in range(1,number+1): result.append(i)
  return sum(result)
```

```python
sum_number(100)
```
